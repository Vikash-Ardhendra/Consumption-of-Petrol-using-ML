```python
df=pd.read_csv('petrol_consumption.csv')
df.head(5)
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
x.head(5)
```

```python
y=df.iloc[:,-1]
y.head(2)
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
std=StandardScaler()
x=std.fit_transform(x)
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
print ('X Shape-',x.shape)
print ('Y Shape-', y.shape)
print ('X_Train Shape-',x_train.shape)
print ('Y_Train Shape-',y_train.shape)
print ('X_Test Shape-',x_test.shape)
print ('Y_Test Shape-',y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt=dt.fit(x,y)
dt
```

```python
y_pred=dt.predict(x)
y_pred
```

```python
accuracy=r2_score(y,y_pred)
print('Accuracy of Decision Tree Regressor is :',accuracy)
```

```python

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
mse
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor( n_estimators=15)
rf=rf.fit(x,y)
```

```python
rf_ypred=rf.predict(x)
accuracy_rf=r2_score(y,rf_ypred)
print('Accuracy of Random forest Regressor is :',accuracy_rf)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,rf_ypred)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
mse= np.square(np.subtract(y,rf_ypred)).mean() 
import math 
RMSE = math.sqrt(mse)
RMSE
```
